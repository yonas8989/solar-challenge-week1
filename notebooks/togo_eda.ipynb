Load the data

```python
import pandas as pd

df = pd.read_csv("../data/togo-dapaong_qc.csv")  # or togo-dapaong.csv
df.head()
df.info()
```

Summary Status and Missing Values

```python
df.describe()

missing = df.isna().sum()
missing[missing > 0]
null_thresh = len(df) * 0.05
```

Outlier Detection and Cleaning useing Z-score Method

```python
from scipy.stats import zscore

cols_to_check = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']
df_z = df[cols_to_check].apply(zscore)

outliers = (df_z.abs() > 3)
outlier_counts = outliers.sum()
print(outlier_counts)
```

Cleaning and Add Cleaning flag

```python
for col in cols_to_check: 
    df[col] = df[col].fillna(df[col].median())
```

```python
df['Cleaning_flag'] = outliers.any(axis=1).astype(int)
```

Save Cleaned Data

```python
df.to_csv("../data/togo_Clean.csv", index=False)
```

Time Series Analysis

```python
import matplotlib.pyplot as plt

df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df.set_index('Timestamp', inplace=True)

df[['GHI', 'DNI', 'DHI', 'Tamb']].plot(figsize=(12, 6))
plt.title('Solar Radiation and Temperature over Time')
plt.ylabel('W/m² or °C')
plt.show()
```

Cleaining Impact 

```python
df.groupby('cleaning_flag')[['ModA', 'ModB']].mean().plot(kind='bar')
plt.title('ModA & ModB Averages Before vs After Cleaning')
plt.show()
```
